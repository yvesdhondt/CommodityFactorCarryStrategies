Shape the market index data in such a way that it is possible to correlate it with the strategy data from this thesis

```python
index_data["Year"] = pd.to_datetime(index_data["Date"]).dt.year
index_data["Month"] = pd.to_datetime(index_data["Date"]).dt.month
index_data["Day"] = pd.to_datetime(index_data["Date"]).dt.day

index_data = index_data.sort_values(by=["Year", "Month", "Day"], ascending=True)
index_data = index_data.groupby(["Year", "Month"]).agg("first").reset_index()

index_data["x"] = index_data["Year"] + (index_data["Month"] - 1) / 12
```

```python
index_data["USBIG"] = index_data["USBIG"] / index_data["USBIG"].shift(1) - 1
index_data["SP500"] = index_data["SP500"] / index_data["SP500"].shift(1) - 1
index_data["RUS2000"] = index_data["RUS2000"] / index_data["RUS2000"].shift(1) - 1

index_data = index_data[["x", "USBIG", "SP500", "RUS2000"]].copy().dropna()
```

```python
index_data
```

```python
# Combine the strategies from the thesis with the market indices
strat_vs_market = regression_table[["x", "Momentum_Carry", "Basis_Momentum_Carry", "Value_Carry"]].copy().merge(
    index_data,
    how="inner"
)
```

```python
strat_vs_market
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix
corr = strat_vs_market.drop(columns=["x"]).corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
fig, _ = plt.subplots(figsize=(10, 9))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, vmin=-1, vmax=1, square=True, linewidths=.5, annot=True, cmap="coolwarm_r")

fig.savefig("strat_vs_market_corr.png", bbox_inches="tight")
```

## Dot-Com Crisis

```python
# Create dataset of cumulative returns for the dot-com period
dot_com_data = strat_vs_market[(strat_vs_market.x >= 1998) & (strat_vs_market.x <= 2003)].copy()

for c in dot_com_data.columns:
    if c != "x":
        dot_com_data[c] = dot_com_data[["x", c]].apply(lambda x: 0 if x[0] == 1998 else x[1], axis=1)

for c in dot_com_data.columns:
    if c != "x":
        dot_com_data[c] = (dot_com_data[c] + 1).cumprod()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

for c in dot_com_data.columns:
    if c != "x":
        ax.plot(dot_com_data["x"], dot_com_data[c], label=c)

ax.set(xlabel='', ylabel='Cumulative Total Return (start at 1)',
       title='Single-Factor Carry Strategies During the Dot-Com Bubble')

ax.grid()

ax.legend()

plt.xticks(
    [1998, 1998.5, 1999, 1999.5, 2000, 2000.5, 2001, 2001.5, 2002, 2002.5, 2003],
    ['Jan, 1998', 'Jul, 1998', 'Jan, 1999', 'Jul, 1999', 'Jan, 2000', 'Jul, 2000', 'Jan, 2001', 'Jul, 2001', 'Jan, 2002', 'Jul, 2002', 'Jan, 2003'],
    rotation=30
)

fig.savefig("dot_com_bubble.png")
plt.show()
```

### Shape the different indices

**USBIG**

```python
# shape usbig
usbig = index_data[["x", "USBIG"]].copy()
usbig.rename(columns={"USBIG": "y"}, inplace=True)
usbig = usbig.merge(risk_free, how="left")

usbig["y_cum"] = (usbig["y"] + 1).cumprod()
usbig["y"] += 1

# Calculate maximum drawdown
usbig["y_cum_max"] = usbig.y_cum.cummax()
usbig["drawdown"] = usbig["y_cum"] / usbig["y_cum_max"] - 1

# Calculate excess returns
usbig["excess_ret"] = usbig["y"] - usbig["rf"]
```

**SP500**

```python
# shape usbig
sp500 = index_data[["x", "SP500"]].copy()
sp500.rename(columns={"SP500": "y"}, inplace=True)
sp500 = sp500.merge(risk_free, how="left")

sp500["y_cum"] = (sp500["y"] + 1).cumprod()
sp500["y"] += 1

# Calculate maximum drawdown
sp500["y_cum_max"] = sp500.y_cum.cummax()
sp500["drawdown"] = sp500["y_cum"] / sp500["y_cum_max"] - 1

# Calculate excess returns
sp500["excess_ret"] = sp500["y"] - sp500["rf"]
```

**RUS2000**

```python
# shape usbig
rus2000 = index_data[["x", "RUS2000"]].copy()
rus2000.rename(columns={"RUS2000": "y"}, inplace=True)
rus2000 = rus2000.merge(risk_free, how="left")

rus2000["y_cum"] = (rus2000["y"] + 1).cumprod()
rus2000["y"] += 1

# Calculate maximum drawdown
rus2000["y_cum_max"] = rus2000.y_cum.cummax()
rus2000["drawdown"] = rus2000["y_cum"] / rus2000["y_cum_max"] - 1

# Calculate excess returns
rus2000["excess_ret"] = rus2000["y"] - rus2000["rf"]
```

### Combine with Table 4

```python
usbig["Strategy"] = "USBIG"
sp500["Strategy"] = "SP500"
rus2000["Strategy"] = "RUS2000"

table_5 = pd.concat([
    table_4,
    usbig[["x", "Strategy", "y", "drawdown", "excess_ret"]],
    sp500[["x", "Strategy", "y", "drawdown", "excess_ret"]],
    rus2000[["x", "Strategy", "y", "drawdown", "excess_ret"]]
])

table_5 = table_5.merge(table_4[["x"]], how="right")
```

```python
from scipy.stats import gmean
```

```python
# Then we aggregate them
table_5.groupby(
    "Strategy"
).agg(
    mean_monthly_return = ("y", lambda x: np.mean(x) - 1),
    mean_geo_monthly_return = ("y", lambda x: gmean(x) - 1),
    monthly_volatility = ("y", np.std),
    monthly_sharpe_ratio = (
        "excess_ret", lambda x: (np.mean(x) - 1) / np.std(x)
    ),
    yearly_sharpe_ratio = (
        "excess_ret", lambda x: np.sqrt(12) * (np.mean(x) - 1) / np.std(x)
    ),
    maximum_drawdown = ("drawdown", np.min)
).transpose()

table_5.groupby(
    "Strategy"
).agg(
    mean_monthly_return = ("y", lambda x: np.mean(x) - 1),
    mean_geo_monthly_return = ("y", lambda x: gmean(x) - 1),
    monthly_volatility = ("y", np.std),
    monthly_sharpe_ratio = (
        "excess_ret", lambda x: (np.mean(x) - 1) / np.std(x)
    ),
    yearly_sharpe_ratio = (
        "excess_ret", lambda x: np.sqrt(12) * (np.mean(x) - 1) / np.std(x)
    ),
    maximum_drawdown = ("drawdown", np.min)
).transpose().to_csv("strategy_vs_index_table.csv", index=False)
```
