```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 4.0, 0.01)
s = 80 + 20 * (t/4) ** 2

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='Maturity (Months)', ylabel='Price',
       title='Futures Curve in Contango')
ax.set_xticks([0,1,2,3,4])
ax.grid()

fig.savefig("test.png")
plt.show()
```
